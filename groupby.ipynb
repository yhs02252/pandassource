#### groupby()

 - groupby("컬럼명") - 기본적으로 컬럼명으로 인식

    - groupby("컬럼명").count() or groupby("컬럼명").agg(function) 등등

```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
mpg_df = sns.load_dataset("mpg")
mpg_df.info()
```

```python
mpg_df
```

```python
# origin 컬럼 기준으로 size()
mpg_df.groupby("origin").size()
```

```python
mpg_df["origin"].value_counts()
```

```python
# origin 컬럼을 기준으로 평균 구하기
mpg_df.groupby("origin")["cylinders"].mean()
mpg_df.groupby("origin")["cylinders"].agg("mean")
mpg_df.groupby("origin")["cylinders"].agg(["mean","count","std"])
```

```python
# origin, model_year 컬럼 기준으로 cylinders 평균

mpg_df.groupby(["model_year","origin"])["cylinders"].mean()
```

```python
mpg_df.describe()
```

```python
df = pd.DataFrame({
    "A" : ["ha","hi","ho","ha","ho"],
    "B" : ["one","two","one","one","two"],
    "Data1" : np.random.randn(5),
    "Data2" : np.random.randn(5)
})

df
```

```python
# A열 기준으로 그룹, sum()
df.groupby("A").sum()
```

```python
df.groupby("A")[["Data1", "Data2"]].agg(["sum","mean", "std"])
```

```python
df.groupby("A")[["Data1", "Data2"]] \
    .agg(["sum","mean", "std"]) \
        .rename(columns={"sum":"합계","mean":"평균","std":"표준편차"})
```

```python
df = pd.DataFrame(
    {
        "School" : ["Yeonhi","Yeonhi","Sungsan","Sungsan","Sungsan"],
        "Name" : ["Heana","Gisu","Una","Naeun","Ziho"],
        "Math" : [92,70,88,92,70]
    }
)

df
```

```python
# school 그룹 / Math 평균 구하기
df.groupby("School")["Math"].mean()
```

```python
# df + df_avg => merge()

df["Avg"] = df.groupby("School")["Math"].transform("mean")
df
```

Apply()
 - () 안에 함수를 적용하여 컬럼을 생성

```python
def grade(num):
    if num > 90:
        grade = "A"
    elif num > 80:
        grade = "B"
    elif num > 70:
        grade = "B"
    else:
        grade = "F"
    return grade
```

```python
df["grade"] = df["Math"].apply(grade)
df
```

```python
dates = {"yyyy-mm-dd":["2005-09-28","2007-10-05","2012-12-20"]}
df = pd.DataFrame(dates)
df
```

```python
# 년도만 잘라서 year 컬럼 생성

def extract_year(col):
    return col.split("-")[0]
```

```python
"2012-12-20".split("-")
```

```python
df["year"] = df["yyyy-mm-dd"].apply(extract_year)
df
```

```python
# 나이 : 2025-2005

def get_age(year, current_year=2025):
    return current_year - int(year)
```

```python
df['age'] = df['year'].apply(get_age)
df
```

```python
# 함수 리턴 : 나는 20살 입니다
def get_introduce(age):
    return f"나는 {age}살 입니다"
```

```python
df['introduce'] = df['age'].apply(get_introduce)
df
```

```python
def get_introduce2(row):
    return f"나는 {row.year}년에 태어났고, 현재 {row.age}살 입니다"
```

```python
df['introduce2'] = df.apply(get_introduce2,axis=1)
df
```
