#### Reshading Data

- Change layout, sorting, reindesing, renaming...
- melt(), pivot(), concat(), sort_values(), rename(), sort_index(), reset_index(), drop()

```python
import pandas as pd
```

```python
df = pd.DataFrame(
    {
        "A" : {
            0:"a",
            1:"b",
            2:"c"
        },
        "B" : {
            0:1,
            1:2,
            2:3
        },
        "C" : {
            0:2,
            1:4,
            2:6
        }
    }
)

df
```

```python
pd.melt(df, id_vars=["A"], value_vars=["B"])
```

```python
pd.melt(df, id_vars=["A"], value_vars=["B","C"])
```

```python
pd.melt(df, value_vars=["A","B","C"])
```

```python
pd.melt(df, value_vars=["A","B","C"]).rename(columns={"variable":"변수","value":"값"})
```

```python
df = pd.DataFrame(
    {
        "foo":["one","one","one","two","two","two"],
        "bar":["A","B","C","A","B","C"],
        "baz":[1,2,3,4,5,6],
        "zoo":["x","y","z","q","w","t"]
    }
)

df
```

```python
df.pivot(index="foo",columns="bar",values="baz")
```

```python
df2 = df.pivot(index="foo",columns="bar",values="baz").reset_index()
df2
```

#### stack, unstack
- stack : 컬럼 => 인덱스로 변경
- unstack : 인덱스 => 컬럼으로 변경

```python
df = pd.DataFrame(
    {
        "A" : [1,3,6,9],
        "B" : [2,4,6,8]
    }, index=pd.MultiIndex.from_tuples([
        ("bar","one"), ("bar","two"), ("baz","one"), ("baz","two")
    ],names=['first','second'])
)
df
```

```python
# 컬럼 => 인덱스
stacked = df.stack()
stacked
```

```python
# 맨 안쪽 인덱스 => 컬럼
stacked.unstack()
```

```python
# 인덱스 번호를 이용해 컬럼으로 변경
stacked.unstack(0)
```

```python
stacked.unstack(1)
```

```python
df = pd.DataFrame({
    "지역" : ["서울","서울","서울","경기","경기","부산","서울","서울","부산","경기","경기","경기"],
    "요일" : ["월요일","화요일","수요일","월요일","화요일","수요일","목요일","금요일","화요일","수요일","목요일","금요일"],
    "강수량" : [100,80,1000,2000,2000,100,50,100,200,100,50,100],
    "강수확률" : [80,70,10,90,60,30,90,20,80,50,10,40]
})

df
```

```python
# 인덱스 변경 : set_index()
df.set_index(["지역", "요일"],inplace=True)
```

```python
# 지역 인덱스를 컬럼으로 변경
df.unstack(0)
```

```python
# 요일 인덱스를 컬럼으로 변경
df.unstack(1)
```

```python
df.unstack(0).stack(0)
```
