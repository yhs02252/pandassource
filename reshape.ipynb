#### Reshading Data

- Change layout, sorting, reindesing, renaming...
- melt(), pivot(), concat(), sort_values(), rename(), sort_index(), reset_index(), drop()

```python
import pandas as pd
```

```python
df = pd.DataFrame(
    {
        "A" : {
            0:"a",
            1:"b",
            2:"c"
        },
        "B" : {
            0:1,
            1:2,
            2:3
        },
        "C" : {
            0:2,
            1:4,
            2:6
        }
    }
)

df
```

```python
pd.melt(df, id_vars=["A"], value_vars=["B"])
```

```python
pd.melt(df, id_vars=["A"], value_vars=["B","C"])
```

```python
pd.melt(df, value_vars=["A","B","C"])
```

```python
pd.melt(df, value_vars=["A","B","C"]).rename(columns={"variable":"변수","value":"값"})
```

```python
df = pd.DataFrame(
    {
        "foo":["one","one","one","two","two","two"],
        "bar":["A","B","C","A","B","C"],
        "baz":[1,2,3,4,5,6],
        "zoo":["x","y","z","q","w","t"]
    }
)

df
```

```python
df.pivot(index="foo",columns="bar",values="baz")
```

```python
df2 = df.pivot(index="foo",columns="bar",values="baz").reset_index()
df2
```

```python
import seaborn as sns

mpg_df = sns.load_dataset("mpg")
mpg_df.head()
```

```python
# mpg 기준으로 정렬

mpg_df.sort_values("mpg")
```

```python
# 내림차순

mpg_df.sort_values("mpg", ascending=False)
```

```python
mpg_df.sort_values()
```

```python
# index 기준으로 정렬

mpg_df.sort_index(ascending=False)
```

```python
# reset_index() : 기존 index는 컬럼으로 변경되고, 0으로 시작하는 새로운 인덱스 생성
#                  series를 df로 변경할 때도 사용

mpg_df.reset_index()
```

```python
mpg_df.drop(columns=["mpg","model_year"])
```
