### Subset(일부분 추출)

 - 행 기준
    - df.head()s, df.tail(), df.sample()
 - 열 기준

```python
import pandas as pd
import numpy as np

# df = pd.DataFrame({
#   "a":[4,5,6,6,np.nan],
#   "b":[7,8,np.nan,9,9], 
#   "c":[10,11,12,np.nan,12], 
# }, index=[1,2,3,4,5])

df = pd.DataFrame({
  "a":[4,5,6,6,np.nan],
  "b":[7,8,np.nan,9,9], 
  "c":[10,11,12,np.nan,12], 
}, index=list(range(1,6)))
df
```

```python
# 일부분 추출 - head(), tail()
df.head()
df.tail()
```

```python
# 임의로 하나 추출

print(df.sample())
print(df.sample(2))
print(df.sample(frac=0.5))
```

```python
# b 컬럼에서 가장 큰 요소가 포함된 하나의 행
df.nlargest(1,"b")
```

```python
# b 컬럼에서 가장 작은 요소가 포함된 두개의 행
df.nsmallest(2,"a")
```

```python
# pip install seaborn

import seaborn as sns

df = sns.load_dataset("iris")
df.head()
```

```python
# 컬럼 추출

df["petal_length"]
```

```python
# 컬럼 추출

df[["petal_length", "sepal_length"]]
# df.petal_length
```

```python
columns = ["petal_length","sepal_width","species"]
df[columns].tail(3)
```

```python
# 행,열 조건
# loc[], iloc[] : 
df.columns
# 모든행, 컬럼('sepal_length', 'sepal_width', 'petal_length') 추출
# df[['sepal_length', 'sepal_width', 'petal_length']]

# df[행, 열] : 행, 열 표현시 ":"
# 시작 : 끝 
df.loc[:,'sepal_length':'petal_length']

# 2~5행, 컬럼('sepal_length','sepal_width','petal_length') 추출
df.loc[2:5,'sepal_length':'petal_length']

# 2~5행, 컬럼('sepal_length','petal_length') 추출
df.loc[2:5,['sepal_length','petal_length']]
```

```python
# iloc : 인덱스 번호 이용
# 맨 뒤 숫자 제외

# 행 제한, 열은 전체(Default)
df.iloc[10:20]

# 행 1,2,4 열 1,3
df.iloc[[1,2,4],[1,3]]
```

```python
# 결과가 [] 조건상 True 인 경우에만 추출
# df['petal_length'] > 3
df[df['petal_length'] > 3]
```

```python
# 행: sepal_length > 5 인 행만 추출하고, 컬럼은 sepal_length, sepal_width 추출
# df.loc[df["sepal_length"]>5, ["sepal_length","sepal_width"]]

# 행: sepal_length > 5 그리고 sepal_length < 6 인 행만 추출하고, 컬럼은 sepal_length, sepal_width 추출
# df.loc[(df.sepal_length>5)&(df.sepal_length<6), ["sepal_length","sepal_width"]]
df.loc[(df["sepal_length"]>5)&(df["sepal_length"]<6), ["sepal_length","sepal_width"]]
```

```python
import pandas as pd

# DataFrame(행, 열)

friend_dict_list = [
    {
        "name":"John",
        "age":25,
        "job":"student"
    },
    {
        "name":"Nate",
        "age":30,
        "job":"teacher"
    },
    {
        "name":"Jenny",
        "age":30,
        "job":"developer"
    }
]

friend_dict_list = pd.DataFrame(friend_dict_list)
friend_dict_list
```

```python
# Nate, Jenny 행만 추출

friend_dict_list.loc[1:]
friend_dict_list.iloc[1:]
```

```python
# Nate, Jenny 행만 추출, name,job 열만 추출
friend_dict_list.loc[[0,2],["name","job"]]
friend_dict_list.iloc[[0,2],[0,2]]
```

```python
# 이름이 Nate이고 나이가 25보다 많은 행 추출
friend_dict_list[(friend_dict_list.name == "Nate") & (friend_dict_list.age > 25)]
```

```python
# query()

friend_dict_list.query("age > 25")
```

```python
# python : and, or, not
# pandas : & , |, !

friend_dict_list.query("age > 25" and "name == 'Nate'")
```
