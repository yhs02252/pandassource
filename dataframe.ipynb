```python
# 삭제
# drop() : 원본에 반영 x
# drop(inplace=True) : 원본에 반영 o
friend_dict_list.drop([0,2], inplace=True)
```

```python
# 컬럼명으로 삭제하기
friend_dict_list.drop(columns="age")
```

```python
# axis - 0: 행, 1: 열
friend_dict_list.drop("age", axis=1)
```

```python
# 컬럼명 수정

friend_dict_list.columns = ["여름","나이","직업"]
friend_dict_list
```
