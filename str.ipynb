```python
# 유튜브 댓글 크롤링 / 데이터프레임

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup

# 크롬 드라이버 : 크롬 웹 브라우저를 동작하기 위해 필요한 프로그램
# selenium 4.6 버전부터 크롬 드라이버 별도 다운로드가 필요 없음

driver = webdriver.Chrome()
# driver.maximize_window()

driver.get("https://www.youtube.com/watch?v=HosW0gulISQ")
time.sleep(5)

# 댓글수집

# 페이지 이동 후 스크롤 이동 : window.scrollTo(시작위치, 종료위치)
# driver.execute_script("window.scrollTo(0, 2160)")

# 1. 현재 문서 높이
prev_height = driver.execute_script("return document.documentElement.scrollHeight")

while True:
    # 현재 가져온 페이지의 높이만큼 스크롤 이동
    driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight)")

    time.sleep(1)

    # 스크롤이 진행된 후 문서 위치
    curr_height = driver.execute_script("return document.documentElement.scrollHeight")

    # 3. 현재 스크롤된 위치와 초기화된 문서의 위치가 같다면 중지
    if curr_height == prev_height:
        break

    # 2. 현재 문서 높이는 스크롤이 진행된 후 높이로 초기화
    prev_height = curr_height

# 맨 처음으로 가기
driver.execute_script("window.scrollTo(0, 0)")

time.sleep(3)

# 댓글 가져오기
response = driver.page_source
soup = BeautifulSoup(response, "html.parser")
```

```python
# 파싱 : 아이디, 코멘트 수집
id_list = soup.select("#author-text > span")
comment_list = soup.select("#content-text > span")
```

```python
for i in range(len(id_list)):
    print(id_list[i].text.strip(), comment_list[i].text.strip())
```

```python
# 엔터 제거 / @ 제거
id_list_clean = []
comment_list_clean = []

for i in range(len(id_list)):
    id, comment = id_list[i].text.strip(), comment_list[i].text.strip()

    clean_id = id.replace("@","")
    clean_id = clean_id.replace("\n","")
    id_list_clean.append(clean_id)

    clean_comment = comment.replace("\n","")
    clean_comment = clean_comment.replace("\t","")
    comment_list_clean.append(clean_comment)

# 확인
for i in range(len(id_list_clean)):
    print(id_list_clean[i], comment_list_clean[i])
```

```python
df = pd.DataFrame(
    {"id" : id_list_clean, "comment" : comment_list_clean}
)
df
```

```python
df.to_csv("./resources/comments.csv", index=False)
```
